```python
import os
from PyPDF2 import PdfReader

# Set the folder path containing your data
pdf_folder = r"C:\Users\saich\PYTHON\GenAI\RAG_OLLAMA\pdf_data"

# List all PDF and TXT files in the folder
pdf_files = [f for f in os.listdir(pdf_folder) if f.lower().endswith('.pdf')]
print(pdf_files)
txt_files = [f for f in os.listdir(pdf_folder) if f.lower().endswith('.txt')]
print(txt_files)

# Read and print the content of each PDF
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() or ""
    print(f"\n--- Content of {pdf_file} ---\n")
    print(text[:1000])  # Print first 1000 characters for preview

# Read and print the content of each TXT file
for txt_file in txt_files:
    txt_path = os.path.join(pdf_folder, txt_file)
    with open(txt_path, 'r', encoding='utf-8') as f:
        text = f.read()
    print(f"\n--- Content of {txt_file} ---\n")
    print(text[:1000])  # Print first 1000 characters for preview
```

```python
from langchain_community.document_loaders import DirectoryLoader,PyPDFLoader, Docx2txtLoader, TextLoader
import os

# Set the folder path containing the documents
pdf_folder = r"C:\Users\saich\PYTHON\GenAI\RAG_OLLAMA\pdf_data"

# Use DirectoryLoader to load all PDFs and DOCX files in the folder
loader = DirectoryLoader(
    pdf_folder,
    glob=["*.pdf", "*.docx"],
    loader_cls=lambda path: PyPDFLoader(path) if path.lower().endswith('.pdf') else Docx2txtLoader(path)
)
documents = loader.load()

# Print the content of each loaded document (first 1000 characters per document)
if not documents:
    print("No PDF or DOCX files found in the folder.")
else:
    for i, doc in enumerate(documents):
        print(f"\n--- Content of Document {i+1}: {doc.metadata.get('source', 'Unknown')} ---\n")
        print(doc.page_content[:1000])
```

```python
import re

def clean_text(text):
    # Remove extra whitespace and newlines
    text = re.sub(r'\s+', ' ', text)
    # Remove non-printable/control characters
    text = re.sub(r'[\x00-\x1f\x7f-\x9f]', '', text)
    # Strip leading/trailing whitespace
    return text.strip()

# Example: Clean and preprocess all loaded document texts
cleaned_documents = []
for i, doc in enumerate(documents):
    raw_text = doc.page_content
    cleaned = clean_text(raw_text)
    cleaned_documents.append(cleaned)
    print(f"\n--- Cleaned Content of Document {i+1}: {doc.metadata.get('source', 'Unknown')} ---\n")
    print(cleaned[:1000])  # Preview first 1000 characters
```

```python
len(cleaned_documents)
```

```python
#pip install langchain-text-splitters

from langchain_text_splitters import (
    CharacterTextSplitter,
    RecursiveCharacterTextSplitter
)
# Example: Use CharacterTextSplitter
char_splitter = CharacterTextSplitter(separator="\n", chunk_size=1000, chunk_overlap=100)
char_chunks = []
for doc in cleaned_documents:
    char_chunks.extend(char_splitter.split_text(doc))
print(f"CharacterTextSplitter produced {len(char_chunks)} chunks.")

# Example: Use RecursiveCharacterTextSplitter
rec_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
rec_chunks = []
for doc in cleaned_documents:
    rec_chunks.extend(rec_splitter.split_text(doc))
print(f"RecursiveCharacterTextSplitter produced {len(rec_chunks)} chunks.")

# Preview a few chunks
for i, chunk in enumerate(rec_chunks[:3]):
    print(f"\n--- Recursive Chunk {i+1} ---\n{chunk[:500]}")
```

```python
'''✅ What to use (LOCAL)

Embeddings: HuggingFace (local model)

Vector DB: FAISS (in-memory or local disk)

LLM (optional): Ollama / LM Studio / skip LLM if just testing retrieval

'''
```

```python
# 1️⃣ Local embeddings (NO API key)

from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# 2️⃣ Create FAISS vector store (LOCAL)

from langchain_community.vectorstores import FAISS

vectorstore = FAISS.from_texts(rec_chunks, embeddings)
```

```python
vectorstore.save_local("faiss_index") # save the FAISS index locally
```

```python
#3️⃣ Test retrieval

# LangChain moved retrievers to the Runnable interface, so .invoke() replaces .get_relevant_documents().

retriever = vectorstore.as_retriever(search_kwargs={"k": 5})

#docs = retriever.invoke("what is BMS?") #get_relevant_documents
docs = retriever.invoke("what is Sensor fusion?")
for d in docs:
    print("\n"+d.page_content)
```

```python
# 1️⃣ Test dataset (you define expected answers)
test_data = [
    {
        "question": "What is Sensor fusion?",
        "answer": "combining data from multiple sensors"
    },
    {
        "question": "What is BMS?",
        "answer": "battery management system"
    }
]


# 2️⃣ Evaluation loop
correct = 0
total = len(test_data)

for item in test_data:
    question = item["question"]
    expected = item["answer"]

    # 🔹 Retrieve context
    docs = retriever.invoke(question)
    context = "\n\n".join([d.page_content for d in docs])

    # 🔹 Prompt
    prompt = f"""
    Answer the question using ONLY the context below.

    Context:
    {context}

    Question:
    {question}
    """

    # 🔹 Model response
    response = llm.invoke(prompt)

    # 🔹 Print results
    print("\n==============================")
    print("Question:", question)
    print("Expected:", expected)
    print("Model:", response)

    # 🔹 Accuracy check (simple)
    if expected.lower() in response.lower():
        print("✅ Correct")
        correct += 1
    else:
        print("❌ Incorrect")


# 3️⃣ Final accuracy
accuracy = correct / total
print("\n==============================")
print(f"Final Accuracy: {accuracy * 100:.2f}%")
```
